```python
#check missing values
df.isnull().sum()
```

```python
# handling missing cvalues in Review_rating column

df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#converting columns names in lower case

df.columns=df.columns.str.lower()
```

```python
# Replaced the space between the columns name with '_'

df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df
```

```python
# create age_group columns

labels=['Young Adult','Adult','Middle aged','Senior']
df['age_group']=pd.qcut(df['age'], q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df.groupby('age_group')['age'].count()
```

```python
# create frequency_of_days of purchasing

frequency_mapping={
    'Fortnightly' : 14,
    'Weekly':7,
    'Monthly':30,
    'Bi-Weekly': 14,
    'Quarterly':90,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
